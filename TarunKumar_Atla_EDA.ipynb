```python
#Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load datasets
customers = pd.read_csv("Customers.csv")
customers
```

```python
products = pd.read_csv("Products.csv")
products
```

```python
transactions = pd.read_csv("Transactions.csv")
transactions
```

```python
# Merge datasets
merged_data = transactions.merge(customers, on="CustomerID").merge(products, on="ProductID")
merged_data
```

```python
# Dataset Overview
print("Dataset Overview:")
print(merged_data.info())
print("\nMissing Values:\n", merged_data.isnull().sum())
print("\nSummary Statistics:\n", merged_data.describe())
```

```python

# Handle missing values (if any)
merged_data.fillna(0, inplace=True)
```

```python
# Univariate Analysis
plt.figure(figsize=(10, 6))
sns.histplot(merged_data["TotalValue"], kde=True, bins=30)
plt.title("Distribution of Total Transaction Value")
plt.show()
```

```python
# Univariate Analysis: Total transaction value by region
plt.figure(figsize=(10, 6))
sns.boxplot(data=merged_data, x="Region", y="TotalValue", palette="viridis")
plt.title("Total Value by Region")
plt.xlabel("Region")
plt.ylabel("Total Transaction Value")
plt.show()
```

```python
# Bivariate Analysis: Quantity vs Total Value grouped by region
plt.figure(figsize=(12, 6))
sns.scatterplot(data=merged_data, x="Quantity", y="TotalValue", hue="Region", alpha=0.7, palette="Set2")
plt.title("Quantity vs Total Value by Region")
plt.xlabel("Quantity")
plt.ylabel("Total Value")
plt.legend(title="Region")
plt.show()
```

```python
# Time-Based Analysis: Transaction trends over months
merged_data["TransactionDate"] = pd.to_datetime(merged_data["TransactionDate"])
merged_data["Month"] = merged_data["TransactionDate"].dt.to_period("M")
monthly_sales = merged_data.groupby("Month")["TotalValue"].sum()
```

```python
# Plot monthly sales trends
plt.figure(figsize=(12, 6))
monthly_sales.plot(kind="bar", color="skyblue")
plt.title("Monthly Sales Trends")
plt.xlabel("Month")
plt.ylabel("Total Sales Value")
plt.xticks(rotation=45)
plt.show()
```

```python
# Business Insights
# 1. Identify the top 10 customers by total spending
top_customers = merged_data.groupby("CustomerID")["TotalValue"].sum().sort_values(ascending=False).head(10)
print("\nTop 10 Customers by Total Spending:\n", top_customers)
# Insight: Top customers contribute significantly to revenue. Focusing on retention strategies for these customers could maximize profitability.
```

```python
# 2. Identify the top 10 products by total sales
top_products = merged_data.groupby("ProductName")["TotalValue"].sum().sort_values(ascending=False).head(10)
print("\nTop 10 Products by Total Sales Value:\n", top_products)
# Insight: Promoting top-selling products and ensuring sufficient stock availability can help sustain and boost sales.
```

```python
# 3. Analyze regional performance by total sales value
regional_performance = merged_data.groupby("Region")["TotalValue"].sum().sort_values(ascending=False)
print("\nRegional Performance (Total Sales Value):\n", regional_performance)
```

```python
# Visualize regional performance
plt.figure(figsize=(10, 6))
regional_performance.plot(kind="bar", color="teal")
plt.title("Sales Value by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales Value")
plt.xticks(rotation=45)
plt.show()
# Insight: Regions with lower sales could benefit from targeted marketing campaigns or localized promotions.
```

```python
# 4. Determine the most popular product categories by sales volume
popular_categories = merged_data.groupby("Category")["Quantity"].sum().sort_values(ascending=False)
print("\nMost Popular Product Categories by Quantity Sold:\n", popular_categories)
```

```python
# Visualize popular product categories
plt.figure(figsize=(10, 6))
popular_categories.plot(kind="bar", color="purple")
plt.title("Most Popular Product Categories by Quantity")
plt.xlabel("Category")
plt.ylabel("Total Quantity Sold")
plt.xticks(rotation=45)
plt.show()
# Insight: Investing in high-demand categories could drive volume growth, while analyzing low-demand categories might uncover opportunities to improve.
```

```python
# 5. Evaluate average spending per customer
avg_spending_per_customer = merged_data.groupby("CustomerID")["TotalValue"].mean().mean()
print(f"\nAverage Spending Per Customer: ${avg_spending_per_customer:.2f}")
# Insight: Understanding the average spending per customer can help set realistic revenue targets and guide strategies to increase spending, such as cross-selling or upselling.
```

```python
#Identify seasonal trends in sales
seasonal_sales = merged_data.groupby(merged_data["TransactionDate"].dt.month)["TotalValue"].sum()
plt.figure(figsize=(10, 6))
seasonal_sales.plot(kind="line", marker="o", color="orange")
plt.title("Seasonal Sales Trends")
plt.xlabel("Month")
plt.ylabel("Total Sales Value")
plt.grid()
plt.show()
# Insight: Understanding seasonal patterns helps optimize inventory and marketing strategies during peak and low seasons.
```
